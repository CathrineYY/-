```python
import pandas as pd

def domestic_info(data):
    province_list = []
    for each in data:
        area = each['area'] # 省份
        confirmed = each['confirmed'] # 累计确诊
        died = each['died'] # 累计死亡
        cured = each['crued'] # 累计治愈
        curConfirm = each['curConfirm'] # 现有确诊
        province_info_all ={'area':area,
                           'confirmed':confirmed,
                           'curConfirm':curConfirm,
                           'died':died,
                           'cured':cured,
                           'country':'中国',
                           'update':'2020-06-23'}
        province_list.append(province_info_all)
    return province_list

# # 省市
def save_province_entity(province_list):
    data = pd.DataFrame({'index:ID': [index + 40 for index in range(len(province_list))],
                         'name': [province['area'] for province in province_list],
                         'update': [province['update'] for province in province_list],
                         'country':'中国',
                         ':LABEL': 'province'
                         })
    data.to_csv(r'./save_province_entity.csv', index=False)

# 累计确诊
def save_confirmed_entity(province_list):
    data = pd.DataFrame({'index:ID': [index + 80 for index in range(len(province_list))],
                         'name': '累计确诊',
                         'number': [province['confirmed'] for province in province_list],
                         'update': [province['update'] for province in province_list],
                         'province':[province['area'] for province in province_list],
                         ':LABEL': 'confirmed'
                         })
#     print(data)
    data.to_csv(r'./save_confirmed_entity.csv', index=False)
    
# 累计死亡
def save_died_entity(province_list):
    data = pd.DataFrame({'index:ID': [index + 120 for index in range(len(province_list))],
                         'name': '累计死亡',
                         'number': [province['died'] for province in province_list],
                         'update': [province['update'] for province in province_list],
                         'province':[province['area'] for province in province_list],
                         ':LABEL': 'died'
                         })
#     print(data)
    data.to_csv(r'./save_died_entity.csv', index=False)
    
# 累计治愈
def save_cured_entity(province_list):
    data = pd.DataFrame({'index:ID': [index + 160 for index in range(len(province_list))],
                         'name': '累计治愈',
                         'number': [province['cured'] for province in province_list],
                         'update': [province['update'] for province in province_list],
                         'province':[province['area'] for province in province_list],
                         ':LABEL': 'cured'
                         })
#     print(data)
    data.to_csv(r'./save_cured_entity.csv', index=False)

def save_china_entity():
    data = pd.DataFrame({'index:ID': [index + 200 for index in range(1)],
                         'name': '中国',
                         'update': '2020-06-12',
                         'continent':'亚洲',
                          ':LABEL': 'country'
                          })
    data.to_csv(r'./save_china_entity.csv', index=False)

    # 现有确诊
def save_curConfirm_entity(province_list):
    data = pd.DataFrame({'index:ID': [index + 240 for index in range(len(province_list))],
                         'name': '现有确诊',
                         'number': [province['curConfirm'] for province in province_list],
                         'update': [province['update'] for province in province_list],
                         'province':[province['area'] for province in province_list],
                         ':LABEL': 'curConfirm'
                         })
#     print(data)
    data.to_csv(r'./save_curConfirm_entity.csv', index=False)
    

def handle_entity(province_list):
    confirmed_list,died_list,cured_list,curConfirm_list = [], [],[],[]
    confirmed_dic,died_dic,cured_dic,curConfirm_dic = {}, {},{},{}

    for elm in province_list:
        confirmed_list.append(elm['confirmed'])
        died_list.append(elm['died'])
        cured_list.append(elm['cured'])
        curConfirm_list.append(elm['curConfirm'])
        
    # 返回的数据格式：“索引值+数据”
    for index, value in enumerate(confirmed_list):
        confirmed_dic[index] = value
    for index, value in enumerate(died_list):
        died_dic[index] = value
    for index, value in enumerate(cured_list):
        cured_dic[index] = value
    for index, value in enumerate(curConfirm_list):
        curConfirm_dic[index] = value
    return confirmed_dic, died_dic,cured_dic,curConfirm_dic

    
def save_province_curConfirm_realtionship(province_list,province_dic,curConfirm_dic):
    data = pd.DataFrame({':START_ID': [
        list(curConfirm_dic.keys())[list(curConfirm_dic.values()).index(info['curConfirm'])] + 240 for info in province_list],
                         ':END_ID': [list(province_dic.keys())[list(province_dic.values()).index(info['area'])] for info
                                     in province_list],
                         'relation': 'belongsTo',
                         ':TYPE': 'belongsTo'
                         })
    data.to_csv(r'./province_curConfirm_realtionship.csv', index=False)
    
def save_province_confirmed_realtionship(province_list,province_dic,confirmed_dic):
    data = pd.DataFrame({':START_ID': [
        list(confirmed_dic.keys())[list(confirmed_dic.values()).index(info['confirmed'])] + 80 for info in province_list],
                         ':END_ID': [list(province_dic.keys())[list(province_dic.values()).index(info['area'])] for info
                                     in province_list],
                         'relation': 'belongsTo',
                         ':TYPE': 'belongsTo'
                         })
    data.to_csv(r'./province_confirmed_realtionship.csv', index=False)
    
def save_province_died_realtionship(province_list,province_dic,died_dic):
    data = pd.DataFrame({':START_ID': [
        list(died_dic.keys())[list(died_dic.values()).index(info['died'])] + 120 for info in province_list],
                         ':END_ID': [list(province_dic.keys())[list(province_dic.values()).index(info['area'])] for info
                                     in province_list],
                         'relation': 'belongsTo',
                         ':TYPE': 'belongsTo'
                         })
    data.to_csv(r'./province_died_realtionship.csv', index=False)

def save_province_cured_realtionship(province_list,province_dic,cured_dic):
    data = pd.DataFrame({':START_ID': [
        list(cured_dic.keys())[list(cured_dic.values()).index(info['cured'])] + 160 for info in province_list],
                         ':END_ID': [list(province_dic.keys())[list(province_dic.values()).index(info['area'])] for info
                                     in province_list],
                         'relation': 'belongsTo',
                         ':TYPE': 'belongsTo'
                         })
    data.to_csv(r'./province_cured_realtionship.csv', index=False)
    
def save_province_country_realtionship(province_list,province_dic,country_dic):
    data = pd.DataFrame({':START_ID': [
        list(province_dic.keys())[list(province_dic.values()).index(info['area'])] for info
                                     in province_list],
        
                         ':END_ID': [list(country_dic.keys())[list(country_dic.values()).index(info['country'])] + 200 for info in province_list],
                         'relation': 'belongsTo',
                         ':TYPE': 'belongsTo'
                         })
    data.to_csv(r'./province_country_realtionship.csv', index=False)    
        
if __name__=='__main__':
    province_list = domestic_info(data)
    save_province_entity(province_list)
    save_curConfirm_entity(province_list)
    save_confirmed_entity(province_list)
    save_died_entity(province_list)
    save_cured_entity(province_list)
    save_china_entity()
    confirmed_dic,died_dic,cured_dic,curConfirm_dic = handle_entity(province_list)
    country_dic={0:'中国'}
    province_dic = {}
    i = 40
    for province in province_list:
        province_dic[i] = province['area']
        i = i+1
    save_province_confirmed_realtionship(province_list,province_dic,confirmed_dic)
    save_province_died_realtionship(province_list,province_dic,died_dic)
    save_province_cured_realtionship(province_list,province_dic,cured_dic)
    save_province_country_realtionship(province_list,province_dic,country_dic)
    save_province_curConfirm_realtionship(province_list,province_dic,curConfirm_dic)
```
